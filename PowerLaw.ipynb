# Power laws
Power law equation:

$
\begin{align}
y & = \alpha \cdot x^{\beta_1} \\
log(y) & = log(\alpha \cdot x^{\beta_1}) \\
  & = log\alpha + logx^{\beta_1} \\
    & = log\alpha + \beta_1logx \\
\end{align}
$






```
import numpy as np
import matplotlib.pyplot as plt
x = range(1,60,3)
x = np.asarray(x, dtype=np.float32)
y = 1.0/x
plt.plot(x,y)
```
